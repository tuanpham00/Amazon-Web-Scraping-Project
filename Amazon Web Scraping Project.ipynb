```python
#import libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
# Connect to website
URL = 'https://www.amazon.fr/JBL-casque-confortable-Bluetooth-autonomie/dp/B0BWVQKSSY/?_encoding=UTF8&pd_rd_w=1fwmd&content-id=amzn1.sym.92118d78-707e-42fb-ae63-d6d321b3cffd&pf_rd_p=92118d78-707e-42fb-ae63-d6d321b3cffd&pf_rd_r=ZM2B5P3BYFN5XF2C704G&pd_rd_wg=eijI5&pd_rd_r=1535e12d-4b88-4f7c-882c-41baa73a88b0&ref_=pd_hp_d_btf_gift_wave2_btf&th=1'
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:146.0) Gecko/20100101 Firefox/146.0"}
page = requests.get(URL, headers=headers)
soup1 = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(),"html.parser")
title = soup2.find(id="productTitle").get_text()
#The are two places that the price appear, so I used select_one to focus just on one link.
price = soup1.select_one("span.a-price.aok-align-center.reinventPricePriceToPayMargin.priceToPay").get_text()
print(title)
print(price)
```

```python
#Clean up the data
price = price.strip()[:-1]
title = title.strip()

print(title)
print(price)
```

```python
import datetime
today = datetime.date.today()
print(today)
```

```python
import csv

header = ['Title', 'Price', 'Date']
data = [title, price, today]

with open('AmazonWebScrapingDataSet.csv', 'w',newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
    
```
